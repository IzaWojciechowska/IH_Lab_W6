```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
from random import randint
```

Scraping the current top 100 songs and their respective artists, and putting the information into a pandas dataframe.

Weekly Top 100 of "hot" songs on Popvortex: http://www.popvortex.com/music/charts/top-100-songs.php

```python
# finding url and store it in a variable
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# downloading html with a get request
response = requests.get(url)
```

```python
response.status_code # 200 status code means OK!
```

```python
# response.content
```

```python
# parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
#print(soup.prettify())
```

```python
# example:
# <div class="chart-content col-xs-12 col-sm-8">
#        <p class="title-artist">
#         <cite class="title">
#          Lil Boo Thang
#         </cite>
#         <em class="artist">
#          Paul Russell
#         </em>
#        </p>
#        <ul>


# finding elements representing items on list
chart_items = soup.select('.chart-content')

songs = []
artists = []

# fetching titles and artists
for item in chart_items:
    title = item.select_one('.title').get_text()
    artist = item.select_one('.artist').get_text()
    songs.append(title)
    artists.append(artist)
    
df1 = pd.DataFrame({"song_title": songs, "artist": artists})


display(df1)
```

# Lab | Web Scraping 

```python
url = "https://www.chartsurfer.de/musik/single-charts-deutschland/jahrescharts/hits-2023/top-100"
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
soup2 = BeautifulSoup(response.content, "html.parser")
print(soup2.prettify())
```

```python
chart_items = soup2.select('.position-relative')

songs = []
artists = []

# fetching names of songs and artists
for item in chart_items:
    title_element = item.select_one('.style-1')
    artist_element = item.select_one('.style-2')

# checking if elements exist
    if title_element and artist_element:
        title = title_element.get_text(strip=True)
        artist = artist_element.get_text(strip=True)
        songs.append(title)
        artists.append(artist)
    
df2 = pd.DataFrame({"song_title": songs, "artist": artists})


display(df2)
```

```python
combined_df = pd.concat([df1, df2], ignore_index=True)

# removing duplicates (song_title,artist)
combined_df.drop_duplicates(subset=['song_title', 'artist'], inplace=True)

# index reset
combined_df.reset_index(drop=True, inplace=True)

display(combined_df)
```
