```python
# collection of dog pictrues

url = 'https://dog.ceo/api/breeds/list/all'

try:
    response = requests.get(url)

    # if the request is successful (code 200)
    if response.status_code == 200:
        data = response.json()
 
        if data and 'message' in data:
            breeds_data = data['message']

            flattened_data = [{'breed': breed, 'sub_breed': sub_breed} 
                              for breed, sub_breeds in breeds_data.items() 
                              for sub_breed in sub_breeds]

            # our dataframe
            df_dog_breeds = pd.DataFrame(flattened_data)

            display(df_dog_breeds.head())
        else:
            print("No dog breeds found")
    else:
        print("Error: ", response.status_code, response.text)

except Exception as e:
    print("Error: ", e)
```

```python
# Cat for every HTTP Status

base_url = "https://http.cat/{}"

# HTTP status codes
status_codes = [
    100, 101, 102, 103, 200, 201, 202, 203, 204, 205, 206, 207, 208, 226,
    300, 301, 302, 303, 304, 305, 307, 308, 400, 401, 402, 403, 404, 405,
    406, 407, 408, 409, 410, 411, 412, 413, 414, 415, 416, 417, 418, 420,
    421, 422, 423, 424, 425, 426, 428, 429, 431, 444, 450, 451, 497, 498,
    499, 500, 501, 502, 503, 504, 506, 507, 508, 509, 510, 511, 521, 522,
    523, 525, 530, 599
]

http_status_data = []

for status_code in status_codes:
    url = base_url.format(status_code)

    try:
        # checking if status code 200
        response = requests.get(url)
        if response.status_code == 200:
            # appending to list
            http_status_data.append({"status_code": status_code, "image_url": url})
        else:
            print("Error: ", response.status_code, response.text)

    except Exception as e:
        print("Error: ", e)

df_status_images = pd.DataFrame(http_status_data)


display(df_status_images)
```
